```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#create a sequence of odd numbers
data = [ x for x in range(1,500,2)] #sequence: 1,3,5, .....499
```

```python
#split data into input (X) and output (y)
X = data[:-1]  #odd numbers sequence
y = data[1:]   #corresponding next digit in sequence
print(X)
print(y)
```

```python
#define the model
model = Sequential()
model.add(Dense(10,input_dim=1))
model.add(Dense(1))
model.summary()
```

```python
#compile the model
model.compile(optimizer=keras.optimizers.Adam(0.01),loss=keras.losses.mean_squared_error)
```

```python
#fit the model
model.fit(X,y,epochs=200,batch_size=10)
```

```python
#make predictions
model.predict([1001])
```

```python
np.round(model.predict([3017,5001]),0)
```
